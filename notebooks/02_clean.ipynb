```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/amazon_sales.csv')
```

```python
df
```

```python
df=df.rename(
    columns={
                'Order ID': 'order_id', 
                'Date': 'date',
                'Status': 'status',
                'Fulfilment': 'fulfilment',
                'Sales Channel ': 'sales_channel',
                'ship-service-level': 'ship_service_level',
                'Style': 'style',
                'SKU': 'sku',
                'Category': 'category',
                'Size': 'size',
                'ASIN': 'asin',
                'Courier Status': 'courier_status', 
                'Qty': 'quantity_ordered',
                'Currency': 'currency',
                'Amount': 'amount',
                'ship-city': 'ship_city',
                'ship-state': 'ship_state',
                'ship-postal-code': 'ship_postal_code',
                'ship-country': 'ship_country',
                'promotion-ids':'promotion_ids',
                'B2B':'b2b',
                'fulfilled-by':'fulfilled_by',        
             }, 
    )

```

```python
df['date']= pd.to_datetime(df['date'], errors='coerce')
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns=['index'])  
```

```python
df['fulfilled_by']= df['fulfilled_by'].fillna('Unknown')
```

```python
df['promotion_ids']= df['promotion_ids'].fillna('None')
```

```python
df['currency'] = df['currency'].fillna('Undefined')
```

```python
df['amount'] = df['amount'].fillna(0)
```

```python
df = df.drop(columns=['Unnamed: 22'])
```

```python
df['ship_city']= df['ship_city'].fillna('Unknown')
```

```python
df['ship_state']= df['ship_state'].fillna('Unknown')
```

```python
df['ship_country']= df['ship_country'].fillna('Unknown')
```

```python
df['ship_postal_code']= df['ship_postal_code'].fillna('Unknown')
```

```python
df['courier_status']= df['courier_status'].fillna('Unknown')
```

```python
df.isna().sum()
```

```python
df.loc[df['status']=='Cancelled', ['courier_status'] ] = 'Cancelled'
```

```python
df.loc[df['status']=='Cancelled', ['courier_status'] ]
```

```python
df['ship_state']=df['ship_state'].str.strip()
```

```python
df['ship_state']=df['ship_state'].str.replace(' ', '_')
```

```python
df['ship_state']=df['ship_state'].str.upper()
```

```python
ship_state_map = {
    "NEW_DELHI": "DELHI",
    "DELHI": "DELHI",
    "ORISSA": "ODISHA",
    "AR": "ARUNACHAL_PRADESH",
    "PONDICHERRY": "PUDUCHERRY",
    "RAJSTHAN": "RAJASTHAN",
    "RAJSH THAN": "RAJASTHAN",
    "RAJSHTHAN": "RAJASTHAN",
    "PB": "PUNJAB",
    "RJ": "RAJASTHAN",
    "NL": "NAGALAND",
    "DADRA_AND_NAGAR": "DADRA_&_NAGAR_HAVELI"
}

df['ship_state'] = df['ship_state'].replace(ship_state_map).str.capitalize()
```

```python
df['ship_state']= df['ship_state'].str.split('/').str[0]
```

```python
df = df[df['ship_state'] != 'APO']
```
